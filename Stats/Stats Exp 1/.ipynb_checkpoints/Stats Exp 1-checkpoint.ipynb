```python
import pandas as pd
import numpy as np
from scipy.stats import trim_mean
from statsmodels import robust
import wquantiles

import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('./IT Salary Survey EU 2018.csv')
```

```python
df.columns
```

```python
df.drop(['Timestamp', 'City', 'Your level', 'Salary one year ago', 'Salary two years ago', 'Are you getting any Stock Options?', 'Company size', 'Company type'], axis=1)
```

```python
df.rename(columns = {'Current Salary':'salary', 'Years of experience':'experience', 'Main language at work':'lang'}, inplace = True)
```

```python
df.fillna(df.median(), inplace=True)
```

```python
df = df.dropna()
df.info()
```

# Estimate of Location
## Location estimate of price and number of reviews


### _Mean_

```python
mean_salary = df['salary'].mean()
trimmed_mean_salary = trim_mean(df['salary'], 0.1)
weighted_mean_salary = np.average(df['salary'], weights=df['experience'])

print(f'Mean price = ${mean_salary}')
print(f'Trimmed mean price = ${trimmed_mean_salary}')
print(f'Weighted mean price = ${weighted_mean_salary}')
```

### _Median_

```python
median_salary = df['salary'].median()
weighted_median_salary = wquantiles.median(df['salary'], weights=df['experience'])

print(f'Median price = ${median_salary}')
print(f'Weighted median price = ${weighted_median_salary}')
```

## Estimates of Variablity


```python
print(f'Standard Deviation = {df["salary"].std()}')
print(f'Median Absolute Deviation = {robust.scale.mad(df["salary"])}')
```

## Percentile and Boxplot

```python
print(df['salary'].quantile([0.05, 0.25, 0.5, 0.75, 0.95]))
```

```python
ax = (df['salary']/1000).plot.box(figsize=(3, 4), color='black',  patch_artist = True, boxprops = dict(facecolor = "#ffae77"), medianprops=dict(color='red'))
ax.set_ylabel('salary (thousands)')

plt.tight_layout()
plt.show()
```

## Frequency table and Histogram

```python
binnedPopulation = pd.cut(df['salary'], 10)
print(binnedPopulation.value_counts())
```

```python
ax = (df["salary"]/1000).plot.hist(figsize=(7, 4), color='#ffae77', edgecolor='black')
ax.set_xlabel('salary (thousands)')

plt.tight_layout()
plt.show()
```

## Density Estimate

```python
ax = df['experience'].plot.hist(density=True, xlim=[0, 12], bins=range(1,12), figsize=(5, 4), color='#ffae77',edgecolor='black')
df['experience'].plot.density(ax=ax, color='red')
ax.set_xlabel('Experience')

plt.tight_layout()
plt.show()
```

## Exploring Binary and Categorical Data

```python
print(df["Gender"].unique())
print(df["lang"].unique())
```

```python

ax = df.boxplot(column='salary', by='Gender', figsize=(5, 4), color='black',  patch_artist = True, boxprops = dict(facecolor = "#ffae77"), medianprops=dict(color='red'))
ax.set_xlabel('Gender')
ax.set_ylabel('Salary')
plt.title('Salary Distribution by Gender')

plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
language_salary_mean = df.groupby('lang')['salary'].mean()

ax = language_salary_mean.plot(kind='bar', figsize=(5, 5), color='#ffae77', edgecolor="black")

ax.set_xlabel('Language')
ax.set_ylabel('Average Salary')
plt.title('Average Salary by Language')

plt.tight_layout()
plt.show()
```

## Scatterplots

```python
ax = df.plot.scatter(x='salary', y='experience', figsize=(6, 4), marker='$\u25EF$', color='#ffa061')
ax.set_xlabel('Salary')
ax.set_ylabel('Experience')
ax.axhline(0, color='grey', lw=1)
ax.axvline(0, color='grey', lw=1)

plt.tight_layout()
plt.show()
```

## Hexagonal binning and Contour

```python
ax = df.plot.hexbin(x='salary', y='Age', gridsize=30, sharex=False, figsize=(5, 4), cmap='Oranges')
ax.set_xlabel('Salary')
ax.set_ylabel('Age')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
sns.kdeplot(data=df, x='salary', y='Age', ax=ax, color='#ffa061')
ax.set_xlabel('Salary')
ax.set_ylabel('Age')

plt.tight_layout()
plt.show()
```

## Categorical and Numeric Data

```python
position_counts = df['Position'].value_counts()
top_5_positions = position_counts.head(5).index
df_top_5_positions = df[df['Position'].isin(top_5_positions)]

ax = df_top_5_positions.boxplot(by='Position', column='salary', figsize=(5, 6), color='black', patch_artist=True, boxprops=dict(facecolor="#ffae77"), medianprops=dict(color='red'))
ax.set_xlabel('')
ax.set_ylabel('Salary')
plt.suptitle('')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.violinplot(data=df, x='lang', y='salary',ax=ax, inner='quartile', color='#ffae77')
ax.set_xlabel('')
ax.set_ylabel('Salary')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
